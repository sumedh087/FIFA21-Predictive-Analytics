```python
##removing the /n pattern in every club name to clean the club column
fifa_data_uncleaned['Club'] = fifa_data_uncleaned['Club'].replace('\n','', regex=True)    
```

```python
fifa_data_uncleaned['Height'].unique()
```

```python
### Function to convert various height formats to centimeters
def convert_to_cm(height_str):
    if 'cm' in height_str:
        return float(height_str.split('cm')[0])
    else:
        feet, inches = map(int, height_str.replace('"', '').split("'"))
        total_inches = feet * 12 + inches
        return total_inches * 2.54
```

```python
### Apply the conversion function to the 'Height' column
fifa_data_uncleaned['Height'] = fifa_data_uncleaned['Height'].apply(convert_to_cm)
```

```python
fifa_data_uncleaned['Height'] = fifa_data_uncleaned['Height'].astype(str) + 'cm'
```

```python
fifa_data_uncleaned['Height'].unique()
```

```python
fifa_data_uncleaned['Weight'].unique()
```

```python
def convert_to_kg(weight_str):
    if 'lbs' in weight_str:
        pounds = float(weight_str.replace('lbs', ''))
        return pounds * 0.453592  # 1 pound is approximately 0.453592 kilograms
    elif 'kg' in weight_str:
        return float(weight_str.replace('kg', ''))
    else:
        # If no suffix is present, assume it's already in kilograms
        return float(weight_str)
```

```python
fifa_data_uncleaned['Weight'] = fifa_data_uncleaned['Weight'].apply(convert_to_kg)
```

```python
fifa_data_uncleaned['Weight'] = fifa_data_uncleaned['Weight'].astype(str) + 'kg'
```

```python
fifa_data_uncleaned['Weight'].unique()
```

```python
fifa_data_uncleaned.columns
```

```python
def currency_data_clean(currency):
    if isinstance(currency,str):
        if currency.find('K') != -1:
            return int(currency.replace('K','').replace('€','')) * 1000
        elif currency.find('M') != -1:
            return int(float(currency.replace('M','').replace('€','')) * 1000000)
    else:
        return currency
```

```python
fifa_data_uncleaned['Value'] = fifa_data_uncleaned['Value'].apply(currency_data_clean).fillna(0).astype(int)
fifa_data_uncleaned['Wage'] = fifa_data_uncleaned['Wage'].apply(currency_data_clean).fillna(0).astype(int)
fifa_data_uncleaned['Release Clause'] = fifa_data_uncleaned['Release Clause'].apply(currency_data_clean).fillna(0).astype(int)

fifa_data_uncleaned['Release Clause'].unique()
```

# Which positions hold the most value according to the players best position and their value?

```python
max_value_players = fifa_data_uncleaned.groupby('Best Position').apply(lambda x: x.loc[x['Over All Rating'].idxmax()])


plt.figure(figsize=(20, 6))
positions = max_value_players['Best Position']
values = max_value_players['Value']
players = max_value_players['Name'] + " " + max_value_players['Over All Rating'].astype(str)
plt.bar(positions, values, label='Players')
plt.xlabel('Position')
plt.ylabel('Value')
plt.title('Player with Maximum Value in Each Position')
plt.xticks(rotation=45)
plt.legend()

# Annotate player names on the bars
for i, value in enumerate(values):
    plt.text(i, value, players.iloc[i], ha='center', va='bottom')

plt.tight_layout()
```
